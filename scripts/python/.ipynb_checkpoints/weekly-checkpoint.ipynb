```python
import nfl_data_py as nfl
import pandas as pd
```

```python
# Extracción de datos desde nfl_data_py
pbp_data = nfl.import_pbp_data([2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022])
```

```python
pd.set_option("display.max_columns", None)
```

```python
# Creación de muestra
data = pbp_data.sample(frac = 0.1, random_state = 1).copy()
data.shape
```

```python
# Lectura de muestra desde archivo csv (En caso de ser necesario)
data = pd.read_csv('C:/Users/gperalta/Downloads/pbp_sample.csv').copy()
```

```python
# Selección de variables importantes
pbp = data[['game_id','old_game_id','home_team','away_team','season_type','week','posteam','posteam_type','defteam',
           'game_date','sp','play_type','yards_gained','pass_length','yards_after_catch','field_goal_result',
           'extra_point_result','two_point_conv_result','td_team','epa','air_epa','yac_epa','comp_air_epa',
           'comp_yac_epa','wpa','vegas_wpa','air_wpa','yac_wpa','comp_air_wpa','comp_yac_wpa','third_down_converted',
           'fourth_down_converted','interception','penalty','fumble_lost','qb_hit','rush_attempt','pass_attempt',
           'sack','touchdown','penalty_team','penalty_yards','season','away_score','home_score','location','result',
           'div_game','roof','surface','success']]
```

```python
pbp.head()
```

```python
weekly = pbp[['game_id', 'home_team', 'away_team', 'season_type', 'season', 'week', 'game_date', 'home_score', 'away_score', 
              'result', 'div_game', 'location', 'roof', 'surface']].drop_duplicates()
weekly
```

```python
game = list(weekly['game_id'])

totales = ['sp', 'yards_gained', 'yards_after_catch', 'epa', 'air_epa', 'yac_epa', 'comp_air_epa', 'comp_yac_epa',
           'wpa', 'vegas_wpa', 'air_wpa', 'yac_wpa', 'comp_air_wpa', 'comp_yac_wpa', 'third_down_converted', 
           'fourth_down_converted', 'interception', 'fumble_lost', 'qb_hit', 'rush_attempt', 'pass_attempt', 
           'sack', 'touchdown', 'success']

castigos = ['penalty', 'penalty_yards']
```

```python
for stat in totales:
    weekly['home_' + stat] = [pbp[(pbp['game_id'] == i) & (pbp['posteam'] == pbp['home_team'])][stat].sum() for i in game]
for cast in castigos:
    weekly['home_' + cast] = [pbp[(pbp['game_id'] == i) & (pbp['penalty_team'] == pbp['home_team'])][stat].sum() for i in game]
for stat in totales:
    weekly['away_' + stat] = [pbp[(pbp['game_id'] == i) & (pbp['posteam'] == pbp['away_team'])][stat].sum() for i in game]
for cast in castigos:
    weekly['away_' + cast] = [pbp[(pbp['game_id'] == i) & (pbp['penalty_team'] == pbp['away_team'])][stat].sum() for i in game]    

weekly.head()
```

```python
pd.set_option("display.max_rows", 70)
weekly.isnull().sum()
```

```python
pd.reset_option("display.max_rows")
weekly['roof'].mode()
```

```python
weekly["roof"] = weekly["roof"].fillna('outdoors')
pd.set_option("display.max_rows", 70)
weekly.isnull().sum()
```

```python
season_dummy = pd.get_dummies(weekly['season_type'])
location_dummy = pd.get_dummies(weekly['location'])
roof_dummy = pd.get_dummies(weekly['roof'])
surface_dummy = pd.get_dummies(weekly['surface'])
```

```python
weekly_totals = pd.concat([weekly, season_dummy, location_dummy, roof_dummy, surface_dummy], axis = 1)
weekly_totals
```

```python
weekly_totals.to_csv('C:/Users/gpera/MCD/TOG/IDI_III_GPH/data_processed/weekly.csv')
pbp.to_csv('C:/Users/gpera/MCD/TOG/IDI_III_GPH/data_raw/pbp_sample.csv')
```
